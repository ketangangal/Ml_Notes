# Histogram , KDE , RUG

```python
df.head()
```

```python
# y axis has no meaning
plt.figure(figsize=(10,4),dpi = 200)
sns.rugplot(x= 'salary',data = df,hue= 'level of education',height = 0.5)
```

```python
# sns.displot() , sns.histplot()
sns.set(style = 'darkgrid')
sns.displot(data = df, x = 'salary' ,bins = 10) # blocks of size 10
```

```python
sns.set(style = 'ticks') # high level cmd ( need to reset kernel)
sns.displot(data = df, x = 'salary' ,bins = 10) # blocks of size 10
```

```python
sns.histplot(data = df,x = 'salary',bins = 10,kde =True) # it doesn't have Rug
```

```python
sns.displot(data = df, x = 'salary' ,kde = True,rug = True) # blocks of size 10
```

```python
sns.kdeplot(data = df,x = 'salary')
```

```python
df = pd.DataFrame(data = sample , columns = ['age'])
```

```python
sns.rugplot(data = df,x = 'age',height = 0.5)
```

```python
sns.displot(data = df,x= 'age',kde = True,rug=True ,color = 'green')
```

```python
plt.figure(figsize=(10,2),dpi = 200)
sns.kdeplot(data = df , x = 'age',clip=[0,100],bw_adjust = 0.5,shade = True) 
# bigger the band width more is the normalization and sometime which is not required
```
