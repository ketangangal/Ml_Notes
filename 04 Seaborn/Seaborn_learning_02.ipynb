```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Distribution among catgegories
* BoxPlot
* Violin Plot
* Swarm Plot
* Boxn Plot ( Latter Value Plot)

```python
df = pd.read_csv("C:\\Users\\Ketan Gangal\\OneDrive\\Desktop\\python_basic\\Ml\\00 ml_course\\05-Seaborn\\StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data= df,y = 'math score',x ='parental level of education',
            hue='test preparation course',palette='Set1')
plt.legend(loc = (1.05,1))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df,x='math score',y ='parental level of education',
              hue='test preparation course',split= True,inner ='quartile')
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df,x='math score',y ='parental level of education',
              hue='test preparation course',split= True,inner ='stick',bw=0.4)
```

```python
plt.figure(figsize = (10,4),dpi=200)
sns.swarmplot(data=df,x = "math score",y='gender',size=3,hue='test preparation course',
             dodge = True)
```

```python
sns.boxenplot(data=df,x='math score',y='gender',
             hue='test preparation course')
```

# Comparison Plot
* joinplot()
* pairplot()

```python
sns.jointplot(data=df,x='math score',y='reading score')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind ='hex')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind ='scatter',alpha=0.5)
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind ='kde',shade=0.3,palette='Set1')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind ='hist',palette='Set1')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',palette='Set1',hue='gender')
```

```python
sns.pairplot(df,hue='gender')
```

```python
sns.pairplot(df,hue='gender',diag_kind='hist',kind='kde',corner = True)
```

```python
g = sns.PairGrid(data=df,hue='gender')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
g.add_legend()
```
