# subplots

```python
x = np.linspace(0,10,11)
y = x**4
```

```python
a = np.arange(1,10)
b = 2*a
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
# axes is numpy array 
# if we assine 2,2 in rows and colmns then we will get a 2d array
# so for that axes takes 2d indexing axes[:1,1:] // axes[0][1]
# fig is canvas on this axes are present [axes1,axes2]

axes[0].plot(x,y)

axes[1].plot(a,b)
```

```python
fig ,axes = plt.subplots(nrows = 2,ncols=2)

axes[0][0].plot(x,y)

axes[0][1].plot(a,b)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)

fig.tight_layout(pad=2) # to adjust spacing between cells automatically 
```

```python
fig ,axes = plt.subplots(nrows = 2,ncols=2,dpi=200)

axes[0][0].plot(x,y)
axes[0][0].set_title('fig 1')

axes[0][1].plot(a,b)
axes[0][1].set_title('fig 2')

axes[1][0].plot(x,y)
axes[1][0].set_title('fig 3')

axes[1][1].plot(a,b)
axes[1][1].set_title('fig 4')

fig.subplots_adjust(wspace=0.5,
    hspace=0.7,left=0.5,
    bottom=0.5)
```

```python
axes.shape
```

```python
axes.diagonal()
```

```python
fig, axes = plt.subplots(3,1)

# if one of rows or col is one than we can loop through it 
for ax in axes:
    ax.plot(x,y)

fig.suptitle("figure level title",X = 0.5,Y=1.1)

fig.tight_layout()
```

# styling and legends

```python
x = np.arange(0,11)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x*2,label='X vs x*2')
axes.plot(x,x*x,label='X vx X*X')
axes.legend(loc=0)
fig.legend(loc=(-0.05,0.5))
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x+1,label='x vx x+1',color = 'green')  # normal word
axes.set_xlim((0,10))

# we can use color words or hex code for custom colors
axes.plot(x,x,label='x vs x',color = '#242930') # custom
axes.set_ylim((0,10))

axes.legend()
axes.grid()

fig.suptitle('Figure 1',X = 0.5,Y= 1.1)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x+1,label='x vx x+1',color = 'green',linewidth=2)  # normal word
axes.set_xlim((0,10))

# we can use color words or hex code for custom colors
axes.plot(x,x,label='x vs x',color = '#242930',lw=2.2) # custom
axes.set_ylim((0,10))

axes.legend()
axes.grid()

fig.suptitle('Figure 1',X = 0.5,Y= 1.1)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x+1,label='x vx x+1',color = 'green',linewidth=2,linestyle = '--')  # normal word
axes.set_xlim((0,10))

# we can use color words or hex code for custom colors
axes.plot(x,x,label='x vs x',color = '#242930',lw=2.2,ls = '-.') # normal style 
axes.set_ylim((0,10))

axes.legend()
axes.grid()

fig.suptitle('Figure 1',X = 0.5,Y= 1.1)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
lines = axes.plot(x,x+1,label='x vx x+1',color = 'green',linewidth=4)  # normal word
lines[0].set_dashes([1,2,2,2])
axes.set_xlim((0,10))

# we can use color words or hex code for custom colors
lines1 = axes.plot(x,x,label='x vs x',color = '#242930',lw=2.5,) # custom style
lines1[0].set_dashes([2,2,5,2,7,2])
axes.set_ylim((0,10))

axes.legend()
axes.grid()

fig.suptitle('Figure 1',X = 0.5,Y= 1.1)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
lines = axes.plot(x,x+1,label='x vx x+1',color = 'green',linewidth=4,linestyle= '-.',
                 marker = 'x',markersize =20)  # normal word
lines[0].set_dashes([1,2,2,2])
axes.set_xlim((0,10))

# we can use color words or hex code for custom colors
lines1 = axes.plot(x,x,label='x vs x',color = '#242930',lw=2.5,marker='o',ms=20) # custom style
lines1[0].set_dashes([2,2,5,2,7,2])
axes.set_ylim((0,10))

axes.legend()
axes.grid()

fig.suptitle('Figure 1',X = 0.5,Y= 1.1)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
lines = axes.plot(x,x+1,label='x vx x+1',color = 'blue',linewidth=4,linestyle= '--',
                 marker = 'o',markersize =20,markerfacecolor = 'red',
                 markeredgewidth = 2,markeredgecolor = 'black')  # normal word
lines[0].set_dashes([1,2,2,2])
axes.set_xlim((0,10))
axes.legend()
axes.grid()

fig.suptitle('Figure 1',X = 0.5,Y= 1.1)
```
