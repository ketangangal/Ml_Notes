# merge and concat

```python
one = {'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']}
one =  pd.DataFrame(one)
```

```python
two = {'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']}
two = pd.DataFrame(two)
```

```python
pd.concat([one,two],axis=1) # feature adding
```

```python
pd.concat([one,two],axis=0) # data point adding
```

```python
two.columns = one.columns
```

```python
d = pd.concat([one,two],axis=0)
d.index = range(len(d))
```

```python
d
```

```python
######################## merge ############################
'''
1. inner 
2. outer merge
3. left or right merge
'''
```

```python
'''pd.merge(
    left,
    right,
    how = 'inner',
    on=None,
    left_on=None,
    right_on=None,
    left_index: bool = False,
    right_index: bool = False)'''
```

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,how='inner',on ='name' )
```

```python
pd.merge(registrations,logins,how = 'outer',on = 'name')
```

```python
pd.merge(registrations,logins,how = 'left',on = 'name')
```

```python
pd.merge(registrations,logins,how = 'right',on = 'name')
```

```python
################### left on Right on left index right index
```

```python
registrations = registrations.set_index('name')
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,how='inner',left_index=True,right_on='name')
```

```python
pd.merge(logins,registrations,right_index=True,left_on='name')
```

```python
registrations = registrations.reset_index()
```

```python
registrations.columns = ['reg_name','reg_id']
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,how ='inner',left_on='reg_name',right_on ='name')
```

```python
registrations.columns = ['name','id']
```

```python
logins.columns = ['id','name']
```

```python
pd.merge(registrations,logins,on='name')
```

```python
pd.merge(registrations,logins,on='name',suffixes=('_reg','_log'))
```
