```python
# apply this function on Rows with column input
def last_four(num):
    # It should return only one value
    return str(num)[-4:]
```

```python
df['Last_cc_digits'] = df['CC Number'].apply(last_four)
```

```python
def hunger_meter(num):
    if num < 10:
        return 'Not Hungery'
    elif num >= 10 and num <40:
        return 'Really Hungery'
    else:
        return 'Hunger Af'
```

```python
df['Hunger'] = df['total_bill'].apply(hunger_meter) 
```

# apply with lambda 

```python
df['total_bill'].apply(lambda x: x*2)
```

lambda x(This x represents input): x*2 (This statement represent output)

```python
df[['total_bill','tip']].head()
```

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return "Comman"
```

```python
df['Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']) , axis = 1)
```

# Vectorize mtd
This is used for convience it does not perform fast but in this perticular case its fast than lambda
it makes a function aware that doesnt aware of numpy, after that that starts acting like numpy and also do broadcast

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```
